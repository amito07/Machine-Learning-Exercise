```python
#set id for the panda series
df = df.set_index('id')
```

```python
# Sub-setting Data
df.columns

#sub set by column name
df = df[['title', 'description', 'publishTime', 'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
df.head()
```

```python
df.shape
```

```python
# Sub set by loc
df_subset1 = df.loc[df['viewCount'] > 1000000]
df_subset1
```

```python
df_subset1.shape
```

```python
#sub set by query
df_subset2 = df_subset1.query('likeCount > 1000000')
df_subset2
```

```python
df_subset2.shape
```

```python
#clean up NAN value
df = df.loc[~df['viewCount'].isna()]
df = df.loc[~df['likeCount'].isna()]
df = df.loc[~df['commentCount'].isna()]
```

```python
# Casting dtypes columns
df['viewCount'] = df['viewCount'].astype('int')
df['likeCount'] = df['likeCount'].astype('int')
df['commentCount'] = df['commentCount'].astype('int')
```

```python
df.dtypes
```

```python
df.head()
```

```python
#creating a new column
df['like_to_view_ratio'] = df['likeCount'] / df['viewCount']
df.head()
```

```python
#Graph Part
df['viewCount'].plot(kind='hist', bins=50, title='Histogram of view Count', figsize=(15, 5))
```

```python
#Scatter Plot
df.plot(kind='scatter', x='viewCount', y='likeCount', title='Scatter Plot of viewCount Vs LikeCount', figsize=(15, 5))
```

```python
#save the new dataset
df.to_csv('New_Data_Set_Mr_Beast_Youtube_Information.csv')
```
